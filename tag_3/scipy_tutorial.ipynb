## Das Scipy Modul
Scipy ist ein mächtiges Python Modul für wissenschaftliches und technisches Rechnen.
Es baut auf Numpy auf und liefert fertige Werkzeuge für komplexe Mathematik, Datenanalyse und Ingenieurarbeiten.

- Numpy: schnelle Arrays und Grundrechnen
- Scipy: fortgeschrittene mathematische Algorithmen
  

```python
import numpy as np
from scipy import integrate, stats
import matplotlib.pyplot as plt

# 1. Messdaten erzeugen: normalverteilt mit Ausreissern
# 2. Mittelwert schaetzen/ Std. mit stats
# 3. Pruefen, ob aus meiner Simulation eine Normalverteilung enstanden sit.
# 4. Ausreisser entfernen

rng = np.random.default_rng(3)
# print(rng)

# 1. Daten erzeugen plus ein paar Ausreisser
x = rng.normal(loc=10.0, scale=2.0, size=200)
# print(x)
# Ein paar Ausreisser selbst hinzufuegen
x[:5] += 15
# type(x)

# 2. Fit einer Normalverteilung
mu, sigma = stats.norm.fit(x)
print(mu, sigma)

# 3. normaltest mit shapiro

W, p = stats.shapiro(x) 
# p > 0.05 normalverteilung 
# p < 0.05 keine normalverteilung
# print("p faktor: (mit Ausreissern) ", p)

# 4. Datenreinigung mit z-score
# scipy -> zscore
z = np.abs(stats.zscore(x))
x_clean = x[z < 3]

# print(x_clean)

mu2, sigma2 = stats.norm.fit(x_clean)
W2, p2 = stats.shapiro(x_clean)

print("Fit (bereinigt): mu =", mu2, "sigma =", sigma2)
print("Shapiro p-value bereinigt:", p2)
```

```python
from scipy import integrate

def f(x):
    return x ** 2

val, err = integrate.quad(f, 0, 3)

print("Integral =", val)
print("Error estimate", err)
```
