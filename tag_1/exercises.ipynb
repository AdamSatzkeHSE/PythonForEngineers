```python
# Übung 1: Kontrollstrukturen

"""
Ausgangssituation:
Ein Datenlogger hat mehrere Messungen von Spannung
und Strom aufgezeichnet. Die elektrische Leistung
soll für jede Messung berechnet werden.
"""


voltages = [5.0, 5.1, 4.9, 5.2]
currents = [0.2, 0.21, 0.19, 0.22]

# # Schreibe hier deinen Code
# for i in range()
for i in range(len(voltages)):
    power = voltages[i] * currents[i]
    print(power) 
```

```python
# Version 2: Erweitert
import random

voltages = []
currents = []

# Aufgabe 1 mit simulierten Werten. Anforderung 5 Messwerte.
for i in range(5):
    voltages.append(random.randint(1, 10))
    currents.append(random.randint(0, 2))

print("Spannungen", voltages)
print("Stromwerte:", currents)
```

```python
# Version 2: Erweitert
# Module random import mit alias 'r'
import random as r

voltages = []
currents = []

# Aufgabe 1 mit simulierten Werten. Anforderung 5 Messwerte.
for i in range(5):
    voltages.append(r.randint(1, 10))
    currents.append(r.randint(0, 2))

print("Spannungen", voltages)
print("Stromwerte:", currents)

print("Zufallspannung:", r.choice(voltages))
```
