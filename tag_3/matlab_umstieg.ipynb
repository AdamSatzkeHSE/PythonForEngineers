### Numpy Tutorial mit Matplotlib Einführung - MATLAB Umstieg Version

Gewöhnliche MATLAB Funktionen in Python umgesetzt


```python
import numpy as np
# Ähnlich zu 0:0.1:10

x = np.arange(0, 10.0 + 1e-12, 0.1)
print(x)
# x = np.linspace(0, 10.0 + 1e-10, 100)
t = np.linspace(0, 2 * np.pi, 400)
print(t)
```

```python
""" Indexing / Slicing / Wertebereiche lesen """

# Indexing
print(x[33])

# Slicing
print(x[1:5])
print(x[-1])
print(x[-3:])
```

```python
""" Plot Basics """
import matplotlib.pyplot as plt

# Berechnungen. Setzen der Werte mit numpy
x = np.linspace(0, 2 * np.pi, 400)
y1 = np.sin(x)
y2 = np.cos(x)

# Visualisierung einstellen
plt.figure()
plt.plot(x, y1, label="sinus")
plt.plot(x, y2, label="cosinus")

# Extras
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.xlabel("x [rad]")
plt.ylabel("Amplitude")
plt.title("Sinus und Cosinus")
# Plotten
plt.show()

```
