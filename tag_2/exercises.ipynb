```python
# Version 3:
# Wir schreiben jetzt eine Funktion, dass die Messergebnisse auf Zwei Nachkommastellen zurückgibt.
# Zuerst muss aber eine Vorabprüfung (Validierung) der Messergebnisse stattfinden.
# Wenn die Leistung > 1.0W ist, soll eine Fehlermeldung angezeigt werden und die Funktion soll abbrechen.
# Ansonsten soll die Liste zurückgegeben werden.

def power_function_3(voltages, currents):
    power = []

    # Validierung 1:
    # Habe ich die gleiche Anzahl an Messwerten für beide Listen?
    if len(voltages) != len(currents):
        print("Fehler bei den Messwerte. Dimensionen stimmen nicht überein!")
        return

    # Länge stimmt, dann mach ich weiter.
    for i in range(len(voltages)):
        # Validierung 2:
        # Ist das Ergebnis > 1.0W ?
        leistung = round(voltages[i] * currents[i], 2)
        # print("Leistung:", leistung)
        if leistung > 1.0:
            print("Fehler: Die Leistung darf nicht > 1.0W sein!")
            print(f"Spannung: {voltages[i]:.2f}, Strom: {currents[i]:.2f}")
            return 
        # Wenn nicht, Leistung hinzufügen. An dieser Stelle ist kein else nötig.
        # Warum?
        power.append(round(voltages[i] * currents[i], 2))

    print("Alles passt!")
    return power

### Main Code ###

print("Messung 1:")
voltages_1 = [5.0, 5.1, 4.9, 5.2]
currents_1 = [0.2, 0.21, 0.19, 0.5]
result = power_function_3(voltages_1, currents_1)
print(result)

print("Messung 2:")
voltages_2 = [4.0, 8.1, 4.9, 5.2]
currents_2 = [0.4, 0.21, 0.19, 0.25]
result = power_function_3(voltages_2, currents_2)
print(result)

print("Messung 3:")
voltages_3 = [3.0, 5.1, 9.9, 5.2]
currents_3 = [0.3, 0.21, 0.19, 0.22]
result = power_function_3(voltages_3, currents_3)
print(result)

print("Messung 4:")
# Hier sollten keine Fehler auftauchen
voltages_4 = [1.0, 4.3, 10, 3.3]
currents_4 = [0.3, 0.21, 0.05, 0.22]
result = power_function_3(voltages_4, currents_4)
print(result)
```

```python
# Wie könnte ich meinen Code schöner machen?
# Eine Funktion für jede Aufgabe.

# Berechnung -- power_function()
# Validierung -- check_power()

def check_power(leistungen):
    """ Validierung der Messergebnissen """

    validated = []
    for power in leistungen:
        if power > 1.0:
            print(f"Power {power} liegt oberhalb der erlaubten Grenze.")
            continue
        else:
            validated.append(round(power, 2))
    return validated

# Wie könnte ich check_power() erweitern, damit ich beliebige Leistungsgrenzen prüfen kann?
def check_power_2(leistungen, grenze):
    validated = []
    for power in leistungen:
        if power > grenze:
            print(f"Power {power:.2f} liegt oberhalb von {grenze:.2f}")
        else:
            validated.append(round(power, 2))
    return validated

def power_function(voltages, currents):
    """ Berechnung der Leistungen. Liste mit Ergebnisse zurueckgeben """
    power = []
    # Wie kann ich jetzt die Listen benutzen?
    for i in range(len(voltages)):
        leistung = voltages[i] * currents[i]
        # Kein return, sondern Information.
        # print(leistung)
        power.append(leistung)
    return power

### Main ###

print("Test check_power()")
voltages_4 = [5.0, 4.3, 10, 3.3]
currents_4 = [0.3, 0.21, 0.05, 0.22]
result = power_function(voltages_4, currents_4)
print(result)
result = check_power(result)
print(result)

print("Test neue Funktion check_power_2() mit eingestellter Grenze")

grenze = 0.75
print("Berechnung ohne Validierung")
result_2 = power_function(voltages_4, currents_4)
print(result_2)
result_2 = check_power_2(result, grenze)
print(result_2)

#
result_3 = power_function(voltages_4, currents_4)
for grenzen in [0.2, 0.3, 0.6, 0.75, 0.99]:
    erlaubt = check_power_2(result_3, grenzen)
    print(erlaubt)
```
