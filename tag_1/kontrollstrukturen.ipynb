## If-Else Anweisungen

```python
# Einfache If-Anweisung

nummer = 10
nummer2 = 0
liste = []
nichts = None

if nummer > 5:
    print("Die Nummer ist größer als 5")

if nummer and liste:
    print("Das wird ausgeführt")
```

```python
# If-Else Anweisung
# Else: Wenn Bedingung im If-Block nicht erfüllt ist, führe das Else-Block aus.
if nummer < 5:
    print("Die Nummer ist kleiner als 5")
else:
    print("Die Nummer ist größer gleich 5")
    
```

```python
# Mehrere Bedingungen mit 

spannung = 24
sp_input = 23.9

if sp_input > spannung:
    print("Device not allowed")
elif 23.0 <= sp_input <= 23.9:
    print("Device Maximum")
elif sp_input == 23:
    print("Device reaching max")
else:
    print("Device allowed")
```

## While-Schleifen in Python

```python
spannung = 10
sp_input = 5

while sp_input <= spannung:
    print("Meine Spannung ist:", sp_input)
    sp_input += 1
```
