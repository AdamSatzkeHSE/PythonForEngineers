## Primitive Datentypen



```python


print("Das ist eine Zeichenkette")
print(17)

print(type(17))
print(type("Python"))

type(print(17))
```

```python
# Das ist ein Kommentar in Python
# Primitive Datentypen
print(type(1.4))
print(type(5))

liste = []
print(type(liste))

print(type(True))
mein_bool = True
mein_bool_2 = False

print(type(mein_bool))

print(type("Hallo Welt"))
```

```python
# Variablen deklarieren

variable_1 = 10
variable_2 = "Hallo Welt"
variable_3 = list()
variable_4 = []

# Variablen aufrufen, verwenden
print(variable_1)
print(variable_2)
print(variable_3)
print(variable_4)
```

```python
# Truthy und Falsey: 
print(bool([]))
print(bool(4))
```

```python
# Vergleichsoperatoren

print(3 > 10)
print(3 < 10)
print(3 == 3)
print(10 != 100)
print(100 >= 99)
print(90 <= 8)
```

```python
print(3 >> 1) # Bitweise Shift Operator -> 3 / 2 hoch 1
print(3 << 2)
# 3: 0000 0011
# 3 >> 1: 0000 0001
# 3 << 2: 0000 1100
print(bin(3))
print(bin(12))

print(hex(3))
print(hex(100))
print(hex(15))
```

```python
print(ord("C"))
print(ord("A"))
print(ord(" "))

print(chr(42))
print(chr(45))
print(chr(32))
```
