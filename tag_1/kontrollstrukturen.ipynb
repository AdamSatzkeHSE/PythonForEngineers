## If-Else Anweisungen

```python
# Einfache If-Anweisung

nummer = 10
nummer2 = 0
liste = []
nichts = None

if nummer > 5:
    print("Die Nummer ist größer als 5")

if nummer and liste:
    print("Das wird ausgeführt")
```

```python
# If-Else Anweisung
# Else: Wenn Bedingung im If-Block nicht erfüllt ist, führe das Else-Block aus.
if nummer < 5:
    print("Die Nummer ist kleiner als 5")
else:
    print("Die Nummer ist größer gleich 5")
    
```

## While-Schleifen in Python

```python
spannung = 10
sp_input = 5

while sp_input <= spannung:
    print("Meine Spannung ist:", sp_input)
    sp_input += 1
```

## For-Schleifen in Python

```python
# for(int i = 0; i < laenge; i++) { 
#     ##
# }
liste = [1, 2, 3]
for element in liste:
    print(element)
    variable = element + 2
    print(variable)
```

```python
# Break and continue in For-Schleifen
liste = [1, 2, 3, 4, "Hallo", "Welt"]

print("Break:")
for i in liste:
    if type(i) is int:
        print(i)
    else:
        break

print("Jetzt bin ich raus")
print(i)

print("Continue:")
for i in liste:
    if type(i) is int:
        continue
    print(i)

```

```python
# Break and continue in While-Schleifen

counter = 0
while True:
    counter += 1
    if counter == 10:
        break
    elif counter == 5:
        print("Sonderbedingung erfüllt:", counter)
        continue
    else:
        print("Normale integers:", counter)

print("Counter hat die 10 erreicht")
```
