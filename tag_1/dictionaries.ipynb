### Dictionaries in Python

```python
# Dictionaries in Python bestehen aus sogenannten Key-Value - Pairs.

coffee = {
    "Espresso": 2,
    "Milchkaffee": 4,
    "Capuccino": "nicht erlaubt"
}

# Zugriff auf Werte (Values) mittels (Schluessel) Key
print(coffee["Espresso"])
print(coffee["Capuccino"])
print(coffee["Milchkaffee"])
```

```python
# Mit Dictionaries arbeiten:

# mit items()
print("Mit items:")
for coffeesort, counter in coffee.items():
    print(f"Key {coffeesort} hat Value {counter}")

# mit keys()
print("Mit Keys:")
for i in coffee.keys():
    print(f"{i}: {coffee[i]}")

# values()
print("Mit values:")
for v in coffee.values():
    print("Value:", v)
```
