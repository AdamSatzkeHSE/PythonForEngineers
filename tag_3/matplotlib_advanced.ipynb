# Matplotlib: sauber, publikationsfähig, professionell

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 400)
y = np.exp(-0.3 * x) * np.sin(2 * x)

# figsize: Kontrolliert Seitenverhältnis
fig, ax = plt.subplots(figsize=(8, 4)) # Breite und Höhe

# Voreinstellungen
ax.plot(x, y, "go--", color="purple", lw=5, alpha=0.9)

ax.plot(x, np.sin(x), color="green")
ax.set_title("Schwingung")
ax.set_xlabel("Zeit [s]", fontsize=12)
ax.set_ylabel("Amplitude", fontsize=12)

# Achsen einstellen
ax.set_xlim(0, 10)
ax.set_ylim(-1, 1)

""" Themes """
plt.style.use("classic")
# Available Themes:
# print(plt.style.available)

max_value = np.argmax(y)
ax.annotate("Maxima",
            xy=(x[max_value], y[max_value]),
            xytext=(x[max_value] + 1, y[max_value] - .5),
            arrowprops=dict(arrowstyle="->"))

# Abstände optimieren
fig.tight_layout()
plt.show()

fig.savefig("schwingung.png") # Best Quality
```
