# OOP: Objektorientierte Programmierung

Programmierparadigma - andere Denkweise zu programmieren, anhand von "Objekten" mit Eigenschaften und Verhalten.

- Eigenschaften: Variablen
- Verhalten: Funktionen


```python
# Klasse: Die Vorlage fuer alle Kaffees.
class Coffee:
    # Konstruktor: Initalisierung der Eigenschaften - Erzeugen eines Objektes
    def __init__(self, wasser, bohnenmenge, milch):
        # Eigenschaften eines Kaffees
        self.wasser = wasser
        self.bohnenmenge = bohnenmenge
        self.milch = milch

    # Verhalten in Funktionen
    def zubereiten(self):
        return f"Kaffee mit {self.wasser} ml Wasser, {self.bohnenmenge} gr Bohnen. Milch: {self.milch}"
    
    def __str__(self):
        return f"Das ist mein Kaffee objekt"
    

coffee1 = Coffee(100, 50, False)
coffee2 = Coffee(40, 100, False)
coffee3 = Coffee(60, 20, False)
coffee4 = Coffee(70, 40, False)

print(coffee1.zubereiten())
print(coffee2.zubereiten())
print(coffee3.zubereiten())
print(coffee4.zubereiten())

print(coffee1)
```

```python
import logging

class MeinLogger:
    def __init__(self, min_level, filename):
        self.min_level = min_level
        self.filename = filename
        logging.basicConfig(
            filename=self.filename,
            level=self.min_level,
            filemode="w"
        )
        
log = MeinLogger("INFO", "test.log")
logging.info("Hallo Welt")
logging.warning("Warning")
```
