## Virtuelle Umgebungen

Eine virtuelle Umgebung ist ein isolierter Python Arbeitsbereich für ein Projekt.

- Eigenes Python Interpreter
- Eigene installierte Pakete
- Unabhängigkeit vom Rest des Systems

### Virtuelle Umgebung definieren

1. Erstellen: python -m venv venv
2. Aktivieren: venv\Scripts\activate
3. Pakete installieren: pip install requests

Nutze venv, wenn du:

Webentwicklung machst (Django, Flask, FastAPI)

kleine bis mittlere Projekte hast

maximale Einfachheit willst

mit pip + PyPI klarkommst

👉 Nutze Anaconda/Conda, wenn du:

Data Science / Machine Learning machst

viele native Libraries brauchst (NumPy, SciPy, PyTorch, etc.)

Probleme mit Abhängigkeiten hattest

reproduzierbare Environments brauchst
