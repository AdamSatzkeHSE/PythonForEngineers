## Das Logging Modul in python

Bisher: print()
logging: Log-Level, Zeitstempel, Ausgabe in Datei, Filterbar

Log-Levels:

- DEBUG: Detraillierte technische Infos
- INFO: Normale Programminformation
- WARNING: Etwas ist nicht 'normal'
- ERROR: Fehler aufgetreten
- CRITICAL: Schwerer Fehler

```python
import logging

logging.basicConfig(level=logging.DEBUG)

logging.debug("Das ist eine DEBUG Nachricht")
logging.info("Programm gestartet")
logging.warning("ACHTUNG!")
logging.error("Fehler aufgetreten")
```

```python
import logging
# Logging in Datei

logging.basicConfig(
    filename="pythonschulung.log",
    level=logging.INFO,
    format="%(asctime)s - %(levelname)s - %(message)s"
)

logging.info("Programmstart")
logging.warning("Etwas stimmt nicht...")
logging.error("Ein Fehler ist augetreten")
logging.info("Programm läuft weiter")
```
