### String-Formatting

```python
# String-Formatting in Python
variable = 11
variable_2 = "Hallo Welt"
print(f"Du bist jetzt eine {variable}: {variable_2}")
print("Du bist jetzt eine", variable, sep='*', end='!')
print("Du bist jetzt eine", variable, ":", variable_2)
```

```python
while True:
    user_input = input("Geben Sie eine Zahl ein: ")
    zahl = float(user_input)
    print(f"Sie haben eine {zahl:.2f} eingegeben")
```
