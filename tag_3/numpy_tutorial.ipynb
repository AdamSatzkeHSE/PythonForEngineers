## Das Numpy Modul - numerisches Python

```python
""" Spannung und Strom sind als Listen gegeben. Die Leistung soll berechnet werden """
voltages = [5.0, 5.1, 4.9]
currents = [0.2, 0.21, 0.19]

powers = []
for i in range(len(voltages)):
    powers.append(voltages[i] * currents[i])

print(powers)

print(voltages)
print(currents)
```

```python
import numpy as np

voltages_np = np.array(voltages)
currents_np = np.array(currents)

print(voltages_np)
print(currents_np)

print(type(voltages_np))
print(type(currents_np))

powers_np = voltages_np * currents_np
print(powers_np)
```

```python
# Ohne numpy

signal = [1, 2, 3, 4]
scaled_signal = []
for i in signal:
    scaled_signal.append(i * 2)

print(scaled_signal)
# Mit numpy
signal_np = np.array(signal)
scaled_signal = signal_np * 2

print(scaled_signal)
```

```python
""" Filtern von Messdaten """
filtered = []
for voltage in voltages:
    if voltage > 5.0:
        filtered.append(voltage)
print(filtered)
# Mit numpy
voltages_np = np.array(voltages)
filtered_np = voltages_np[voltages_np > 5.0]

print(filtered_np)
```

```python
""" Mittelwert berechnen """

values = [1, 2, 3, 4]

mittelwert = sum(values) / len(values)
print(mittelwert)

# Mit numpy
values_np = np.array(values)
# values_np = np.array([.1, .2, .3])
np.array
mittelwert_np = np.mean(values_np)

print(mittelwert_np)
```

```python
""" Zeitachse für Simulation """
t = []
step = 0.1
value = 0

while value <= 10:
    t.append(round(value, 2))
    value += step

print(t)

# Mit numpy
t_np = np.linspace(0, 10, 100)
print(t_np)
```

```python
""" Mathematische Funktionen """

import math

values = [0, 1, 2, 3]
result = []
for value in values:
    result.append(math.sin(value))

print(result)

# Mit numpy
values_np = np.array(values)
result_np = np.sin(values_np)

print(result_np)
```
