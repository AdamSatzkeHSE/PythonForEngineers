## Datenstrukturen

```python
# Listen
liste = [True, 23, 9.4]

print(liste)

for i in liste:
    print(i)


print(liste[0])
print(liste[1])
# print(liste[10])

# Indexing - Zuweisung

liste[1] = "Python"
print(liste)
print(type(liste))
```

```python
# Tuples

mein_tuple = (1, 2, 3, 3, 3, 3)
print(mein_tuple)
print(type(mein_tuple))

# mein_tuple[0] = 10
print(mein_tuple.count(3))
# mein_tuple.remove(3) Gibts nicht, da tuple unveränderbar sind.
```
