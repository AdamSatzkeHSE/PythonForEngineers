| Zweck               | Syntax           | Beispiel                       | Ergebnis               |
| ------------------- | ---------------- | ------------------------------ | ---------------------- |
| **Überschrift 1–6** | `#` bis `######` | `## Titel`                     | ## Titel               |
| **Fett**            | `**Text**`       | `**wichtig**`                  | **wichtig**            |
| *Kursiv*            | `*Text*`         | `*Hinweis*`                    | *Hinweis*              |
| ~~Durchgestrichen~~ | `~~Text~~`       | `~~alt~~`                      | ~~alt~~                |
| Inline-Code         | `` `code` ``     | `` `print()` ``                | `print()`              |
| Codeblock           | ` `              | `python<br>print("Hi")<br>`    | Formatierter Codeblock |
| Aufzählung          | `-` oder `*`     | `- Punkt`                      | • Punkt                |
| Nummerierte Liste   | `1.`             | `1. Schritt`                   | 1. Schritt             |
| Link                | `[Text](URL)`    | `[Google](https://google.com)` | klickbarer Link        |
| Bild                | `![Alt](URL)`    | `![Bild](url)`                 | Bildanzeige            |
| Zitat               | `>`              | `> Hinweis`                    | eingerückter Text      |
| Horizontale Linie   | `---`            | `---`                          | Trennlinie             |
| Tabelle             | `\|` + `---`     | siehe unten                    | Tabelle                |
| Zeilenumbruch       | zwei Leerzeichen | `Text  `                       | neue Zeile             |
